```python
import tiktoken
import os
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)

model = os.getenv("OPENAI_MODEL")
api_key = os.getenv("OPENAI_API_KEY")
```

```python
encoding = tiktoken.encoding_for_model(model)

tokens = encoding.encode("Hello, mon nom c'est Jordan MD")
```

```python
tokens
```

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} : {token_text}")
```

## Verification de la conformite

```python
from openai import OpenAI

client = OpenAI()
```

# Le message envoyer a OpenAI est une liste

```python
messages = [
    {"role": "system", "content": "Tu est un assistant qui aide"},
    {"role": "user", "content": "Quel est mon nom?"}
]
```

```python
respone = client.chat.completions.create(
    model=model,
    messages=messages
)
respone.choices[0].message.content
```
