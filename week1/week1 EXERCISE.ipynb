## Correction de l'exercice avec Gemini et Ollama

### Commencons par ollama avec le modele llama3.2:1b

```python
# Initialiser l'objet qui fera le lien avec les endpoint de gpt
ollama_client = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")
```

```python
system_prompt_ollama =" Tu es un assistant specialiser dans la pedagogie, \
qui s'exprime de facon clair, comprhensible et precise pour ce diriger son objet> \
Tu es un maitre de Python. \
Et repond au fromat markdown."
```

```python
def user_prompt_ollama (question):
    user_prompt_ollama = f"Tu vois cette question:\n {question}"
    user_prompt_ollama += "Tu es va repondre a cette question en francais"
    user_prompt_ollama += "Et tu va tenir compte de mon interret pour le developpement backend et IA \
pour me fonir une explication et la valeur que j'avais un comphrension total de cette notion "
    return user_prompt_ollama
```

```python
print(user_prompt_ollama(question))
```

```python
def get_message_ollama(question):
    return [
        {"role": "system", "content": system_prompt_ollama},
        {"role": "user", "content": user_prompt_ollama(question)}
    ]
```

```python
print(get_message_ollama(question))
```

```python
def get_answer(question):
    response = ollama_client.chat.completions.create(
        model = MODEL_LLAMA,
        messages = get_message_ollama(question)
    )
    
    resultat = response.choices[0].message.content
    return resultat
```

```python
get_answer(question)
```

```python
def display_answer(question):
    answer = get_answer(question)
    display(Markdown(answer))
    
```

```python
display_answer(question)
```

### Et maintenant Gemini avec le modele Gemini 2.0 flash

```python
# Load environment variables in a file called .env

load_dotenv(override=True)
api_key = os.getenv('GEMINI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
# Configure Gemini API with the new client
client = genai.Client(api_key=api_key)

# Models are how you specify which version of Gemini you want to use.
model = 'gemini-2.0-flash'
```

```python
system_prompt_gemini =" Tu es un assistant specialiser dans la pedagogie, \
qui s'exprime de facon clair, comprhensible et precise pour ce diriger son objet> \
Tu es un maitre de Python. \
Et repond au fromat markdown."

def user_prompt_gemini (question):
    user_prompt_gemini = f"Tu vois cette question:\n {question}"
    user_prompt_gemini += "Tu es va repondre a cette question en francais"
    user_prompt_gemini += "Et tu va tenir compte de mon interret pour le developpement backend et IA \
pour me fonir une explication et la valeur que j'avais un comphrension total de cette notion "
    return user_prompt_gemini
```

```python
user_prompt_ollama(question)
```

```python
def get_answer_gemini(question):
    response = client.models.generate_content(
        model=MODEL_GEMINI,
        contents=user_prompt_gemini(question),
        config=types.GenerateContentConfig(
            system_instruction=system_prompt_gemini,
            temperature=0.7
        )
    )
    
    return response
```

```python
get_answer_gemini(question)
```

```python
def display_answer_gemini(question):
    resultat = get_answer_gemini(question)
    display(Markdown(resultat))
```
