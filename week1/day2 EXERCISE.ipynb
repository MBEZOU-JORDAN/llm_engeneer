# NOW the exercise for you

Take the code from day1 and incorporate it here, to build a website summarizer that uses Llama 3.2 running locally instead of OpenAI; use either of the above approaches.

```python
import ollama
import requests
from bs4 import BeautifulSoup
from IPython.display import Markdown, display
```

```python
# Constants/ Model
MODEL = "llama3.2:1b"
```

```python
class Website:
    url: str
    title: str
    texte: str
    
    def __init__(self, url):
        self.url = url
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        self.title = soup.title.string if soup.title else "No title found"
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        self.texte = soup.body.get_text(separator="\n", strip=True) if soup.body else "No body text found"
```

```python
ed = Website("https://en.wikipedia.org/wiki/Artificial_intelligence")
print(ed.title)
print(ed.texte)
```

```python
# Prompts pour le LLM
system_prompt = """
Tu es un expert en IA et partuiculierrement en LLM 
Et aussi un assistaant pédagogique pour les étudiants qui veulent apprendre à utiliser les LLMs
"""

def user_prompt(website):    
    user_prompt = f"Tu voici le site de titre {website.title}"
    user_prompt += """Le contenu du site est le suivant; \
spt analyse et fait un bref resume du site en markdown. \
ressortir les information les plus importante et les plus pertinentes sans oublier de parles des actualites. \
"""
    user_prompt += "Voici le contenu du site: \n\n"
    user_prompt += website.texte
    return user_prompt
```

```python
# Reunir les prompt ey le mettre dans le format adequat pour ollama 
def message_for_ollama(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt(website)}
    ]
```

```python
# Faire le resume du site web en utilisant le LLM local
def summarize(url):
    website = Website(url)
    messages = message_for_ollama(website)
    response = ollama.chat(model=MODEL, messages=messages)
    return response['message']['content']
```

```python
# Teste de la fonction de resumer de site 
summarize("https://en.wikipedia.org/wiki/Artificial_intelligence")
```

```python
# Un fonction de afficher le resultat du llm de facin plus presentable
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```

```python
# Tester de l'afficher
display_summary("https://en.wikipedia.org/wiki/Artificial_intelligence")   
```
